```python
def track_item_boxes(imgpath1, imgpath2, img1boxclasslist, visualize=True):
    # Create temp directory for tracking
    tempfolder = "./tempdir"
    if not os.path.exists(tempfolder):
        os.mkdir(tempfolder)

    # Copy images to the temp directory
    shutil.copy(imgpath1, tempfolder + "/00000.jpg")
    shutil.copy(imgpath2, tempfolder + "/00001.jpg")

    # Load and display the first image with the bounding box
    fig, ax = plt.subplots()
    plt.title("Original Image with Bounding Box")
    ax.imshow(Image.open(tempfolder + "/00000.jpg"))
    rect = patches.Rectangle((img1boxclasslist[0][0][0], img1boxclasslist[0][0][2]),
                             img1boxclasslist[0][0][1] - img1boxclasslist[0][0][0],
                             img1boxclasslist[0][0][3] - img1boxclasslist[0][0][2],
                             linewidth=2, edgecolor='green', facecolor='none')
    ax.add_patch(rect)
    plt.show()

    # Display the second image with a mask
    plt.figure(figsize=(6, 4))
    plt.title("Tracked Object in Second Image")
    plt.imshow(Image.open(tempfolder + "/00001.jpg"))
    rect2 = patches.Rectangle((img1boxclasslist[0][0][0], img1boxclasslist[0][0][2]),
                             img1boxclasslist[0][0][1] - img1boxclasslist[0][0][0],
                             img1boxclasslist[0][0][3] - img1boxclasslist[0][0][2],
                             linewidth=2, edgecolor='red', facecolor='none')
    plt.gca().add_patch(rect2)
    plt.show()
```
